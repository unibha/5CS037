```python
#problem:2
#task:1
medical_df = pd.read_csv("/content/drive/MyDrive/medical_students_dataset.csv")
medical_df.head()
```

```python
#task:2
medical_df.info()
null_counts = medical_df.isnull().sum()
print("\nNull values per column:")
print(null_counts)
missing_cols = null_counts[null_counts > 0].index
print("\nColumns with missing values:")
print(missing_cols)
```

```python
#task:3
numeric_cols = medical_df.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = medical_df.select_dtypes(include=["object"]).columns
for col in numeric_cols:
    missing = medical_df[col].isnull().sum()
    if missing > 0:
        mean_value = medical_df[col].mean()
        medical_df[col] = medical_df[col].fillna(mean_value)
        print(f"Filled {missing} missing values in numeric column '{col}' with mean = {mean_value:.2f}")
    else:
        print(f"No missing values in numeric column '{col}'")
for col in categorical_cols:
    missing = medical_df[col].isnull().sum()
    if missing > 0:
        mode_value = medical_df[col].mode()[0]
        medical_df[col] = medical_df[col].fillna(mode_value)
        print(f"Filled {missing} missing values in categorical column '{col}' with mode = {mode_value}")
    else:
        print(f"No missing values in categorical column '{col}'")
```

```python
#task:4
dup_count = medical_df.duplicated().sum()
print("Number of duplicated rows:", dup_count)
medical_df_clean = medical_df.drop_duplicates().reset_index(drop=True)
print("Shape before:", medical_df.shape)
print("Shape after removing duplicates:", medical_df_clean.shape)
```

```python
#task:3.2
titanic_df = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")
titanic_df.head()
titanic_df.info()
```

```python
#problem:1
cols = ["Name", "Pclass", "Sex", "Age", "Fare", "Survived"]
titanic_sub = titanic_df[cols]
first_class = titanic_sub[titanic_sub["Pclass"] == 1]
fare_mean = first_class["Fare"].mean()
fare_median = first_class["Fare"].median()
fare_max = first_class["Fare"].max()
fare_min = first_class["Fare"].min()
print("First-class Fare statistics:")
print(f"Mean   : {fare_mean:.2f}")
print(f"Median : {fare_median:.2f}")
print(f"Max    : {fare_max:.2f}")
print(f"Min    : {fare_min:.2f}")
```

```python
#problem:2
age_null_count = first_class["Age"].isnull().sum()
print("Number of null values in Age (first-class subset):", age_null_count)
first_class_no_age_null = first_class.dropna(subset=["Age"])
print("Shape before dropping:", first_class.shape)
print("Shape after dropping:", first_class_no_age_null.shape)
```

```python
#problem:3
embarked_dummies = pd.get_dummies(titanic_df["Embarked"], prefix="Embarked")
titanic_encoded = pd.concat([titanic_df, embarked_dummies], axis=1)
titanic_encoded = titanic_encoded.drop(columns=["Embarked"])
titanic_encoded.head()
```

```python
#problem:4
import matplotlib.pyplot as plt
survival_by_sex = titanic_df.groupby("Sex")["Survived"].mean()
print(survival_by_sex)
plt.figure(figsize=(6, 4))
survival_by_sex.plot(kind="bar")
plt.title("Mean Survival Rate by Sex")
plt.ylabel("Mean of Survived")
plt.xlabel("Sex")
plt.xticks(rotation=0)
plt.ylim(0, 1)
plt.show()
```

```python
#problem:5
titanic_port = titanic_df.dropna(subset=["Embarked", "Sex", "Survived"])
survival_by_sex_embarked = (
    titanic_port
    .groupby(["Embarked", "Sex"])["Survived"]
    .mean()
    .unstack()
)

print(survival_by_sex_embarked)
survival_by_sex_embarked = survival_by_sex_embarked.loc[["C", "Q", "S"]]
ax = survival_by_sex_embarked.plot(kind="bar", figsize=(8, 5))
plt.title("Mean Survival Rate by Sex and Port of Embarkation")
plt.ylabel("Mean of Survived")
plt.xlabel("Port of Embarkation")
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.legend(title="Sex")
plt.show()
```
