**Recursion Exercises**

**Task 1 - Sum of Nested Lists**

```python
def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total

nested_list_sample = [1, [2, [3, 4], 5], 6, [7, 8]]
print('nested_list:', nested_list_sample)
print('sum:', sum_nested_list(nested_list_sample))
```

**Task 2 - Generate Permutations (unique) recursively**

```python
def generate_permutations(s):
    results = []
    def helper(prefix, remaining):
        if not remaining:
            results.append(prefix)
            return
        seen = set()
        for i, ch in enumerate(remaining):
            if ch in seen:
                continue
            seen.add(ch)
            helper(prefix + ch, remaining[:i] + remaining[i+1:])
    helper('', s)
    return results

print(generate_permutations('abc'))
print(generate_permutations('aab'))
```

**Task 3 - Directory Size Calculation**

```python
def calculate_directory_size(directory):
    total = 0
    for name, value in directory.items():
        if isinstance(value, dict):
            total += calculate_directory_size(value)
        else:
            total += value
    return total

directory_structure = {
    'file1.txt': 200,
    'file2.txt': 300,
    'subdir1': {
        'file3.txt': 400,
        'file4.txt': 100
    },
    'subdir2': {
        'subsubdir1': {
            'file5.txt': 250
        },
        'file6.txt': 150
    }
}

print('Total directory size (KB):', calculate_directory_size(directory_structure))
```

**Dynamic Programming Exercises**

**Task 1 - Coin Change**

```python
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i-coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1

print(min_coins([1,2,5], 11))
```

**Task 2 - Longest Common Subsequence (LCS)**





```python
def longest_common_subsequence(s1, s2):
    n, m = len(s1), len(s2)
    dp = [[0]*(m+1) for _ in range(n+1)]
    for i in range(1, n+1):
        for j in range(1, m+1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    return dp[n][m]

print(longest_common_subsequence('abcde', 'ace'))
```

**Task 3 - 0/1 Knapsack Problem:**

```python
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [0] * (capacity + 1)

    for i in range(n):
        for w in range(capacity, weights[i] - 1, -1):

            dp[w] = max(dp[w], dp[w - weights[i]] + values[i])

    return dp[capacity]
weights = [1, 3, 4, 5]
values  = [1, 4, 5, 7]
capacity = 7

print("Maximum value:", knapsack(weights, values, capacity))
```
